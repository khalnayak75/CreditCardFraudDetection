# Importing CSV

```python
import pandas as pd
ccfr= pd.read_csv('creditcard.csv')
ccfr.head()
```

```python
import matplotlib.pyplot as plt
x= ccfr['Class']
plt.figure(figsize=(5,7))
plt.hist(x,bins=2)
plt.xticks(ticks=[0,1])
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
fraud_df= ccfr[ccfr['Class']==1]
nonfraud_df= ccfr[ccfr['Class']==0]
fraudcount= fraud_df.count()[0]
nonfraudcount= nonfraud_df.count()[0]
print ("Frauds=",fraudcount)
print ("Non Frauds=",nonfraudcount)
```

# Visualization

```python
import numpy as np
from scipy import stats
import seaborn as sns
for c in ccfr.columns[0:30]:
    print ("******************** COLUMN ",c," ***********************")
    col= ccfr[c]
    col=np.array(col)
    col_mean= np.mean(col)
    col_median= np.median(col)
    col_mode= stats.mode(col)[0][0]
    col_std= np.std(col)
    col_var= np.var(col)
    col_range= col.max()-col.min()
    fig=sns.FacetGrid(ccfr,hue="Class",height=5,aspect=2,palette=["blue", "green"])
    fig.map(sns.distplot,c)
    fig.add_legend(labels=['Non Fraud','Fraud'])
    plt.axvline(col_mean,color='red',label='mean')
    plt.axvline(col_median,color='yellow',label='median')
    plt.axvline(col_mode,color='black',label='mode')
    plt.legend()
    plt.show()
    print ("Standard Dev.=",col_std)
    print ("Variance.=",col_var)
    print ("Range=",col_range)
    print (" ")
```

# Principal Component Analysis using StandardScaler

```python
#STANDARSIZATION using MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
ccfrcopy= ccfrnew.copy()
ccfrcopy= ccfrcopy.drop("Class", axis= 1)
std_scale = StandardScaler().fit(ccfrcopy)
ccfrpca = std_scale.transform(ccfrcopy)
ccfrpca
```

```python
#deciding n_components
pca= PCA().fit(ccfrpca)
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Variance (%)') #for each component
plt.title('Credit Card Dataset Explained Variance')
plt.grid()
plt.show()
```

# Principal Component Analysis using Z Scores (user def fn)

```python
def zscore(d):
    return (d-np.mean(ccfrcopy[col]))/np.std(ccfrcopy[col])
ccfrcopy= ccfrnew.copy()
ccfrcopy= ccfrcopy.drop("Class", axis= 1)
for col in ccfrcopy.columns:
    ccfrcopy[col]= ccfrcopy[col].apply(zscore)
```

```python
#to show how Z score acts as a standard, hence not changing the original correlation or covariance
plt.figure(figsize=(20,20))
sns.heatmap(ccfrcopy.corr(), cmap="coolwarm", annot=True)
```

```python
#deciding n_components
pca= PCA().fit(ccfrcopy)
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Variance (%)') #for each component
plt.title('Credit Card Dataset Explained Variance')
plt.grid()
plt.show()
```

```python
ccfrcopy= ccfrnew.copy()
ccfrcopy= ccfrcopy.drop("Class", axis= 1)
pca= PCA(n_components=18)
componentdf= pca.fit_transform(ccfrcopy)
componentdf= pd.DataFrame(componentdf)
ccfrnew.reset_index(drop=True,inplace= True)
componentdf= componentdf.join(ccfrnew['Class'])
componentdf
```

```python
X = componentdf.drop('Class',axis=1) #independent variables columns Time to Amount
y = componentdf['Class'] #dependent varibles i.e. Class to be predicted
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size= 0.4, random_state=0) #60:40 Split
print ("Training Data:")
print ("xtrain:")
print (xtrain.head())
print ("xtest:")
print (xtest.head())
print ("Test Data:")
print ("ytrain:")
print (ytrain.head())
print ("ytest:")
print (ytest.head())
print (xtrain.shape,xtest.shape,ytrain.shape,ytest.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
k_range= range(5,9)
scores= []
for k in k_range:
    knn= KNeighborsClassifier(n_neighbors=k)
    knn.fit(xtrain, ytrain)
    scores.append(knn.score(xtest,ytest))
    print("k=",k,"; Score=",knn.score(xtest, ytest))
plt.figure(figsize= (6,4))
plt.scatter(k_range,scores)
plt.xticks([5,6,7,8])
plt.xlabel("k Value")
plt.ylabel("Scores")
plt.title("K vs Score Plot")
plt.show()
```
